<h1>Descriptive statistics<h1>

```python
print("Mean Age:", df["Age"].mean())
print("Mean Weight:", df["Weight_kg"].mean())
```

```python
#median
print("Median Age:", df["Age"].median())
print("Median Weight:", df["Weight_kg"].median())
```

```python
#Mode
print("Mode Age:", df["Age"].mode()[0])
print("Mode Weight:", df["Weight_kg"].mode()[0])
```

```python
#minimum and maximmumm
print("Minimum Age:", df["Age"].min())
print("Minimum Weight:", df["Weight_kg"].min())
print("Maximum Age:", df["Age"].max())
print("Maximum Weight:", df["Weight_kg"].max())
```

```python
#range
age_range = df["Age"].max() - df["Age"].min()
weight_range = df["Weight_kg"].max() - df["Weight_kg"].min()
print("Age Range:", age_range)
print("Weight Range:", weight_range)
```

```python
#variance
print("Age Variance:", df["Age"].var())
print("Weight Variance:", df["Weight_kg"].var())
```

```python
#standard deviation
print("Age Std:", df["Age"].std())
print("Weight Std:", df["Weight_kg"].std())
```

```python
#Quatriles
print(df["Age"].quantile([0.25, 0.50, 0.75]))
print(df["Weight_kg"].quantile([0.25, 0.50, 0.75]))
```

```python
#percentiles
print(df["Age"].quantile([0.10, 0.25, 0.50, 0.75, 0.90]))
print(df["Weight_kg"].quantile([0.10, 0.25, 0.50, 0.75, 0.90]))
```

```python
#skewness
print("Age Skewness:", df["Age"].skew())
print("Weight Skewness:", df["Weight_kg"].skew())
```

```python
                                                        #Histogram
#Age
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Weight
plt.figure(figsize=(8,5))
plt.hist(df["Weight_kg"], bins=10)
plt.title("Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.show()
```

```python
                                       #Distribution Plot (Histogram + KDE)
#Age
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()

#Weight
plt.figure(figsize=(8,5))
sns.histplot(df["Weight_kg"], kde=True)
plt.title("Weight Distribution")
plt.show()
```

```python
                                                   #BoxPlot
#Age
plt.figure(figsize=(6,4))
plt.boxplot(df["Age"])
plt.title("Box Plot of Age")
plt.show()

#weight
plt.figure(figsize=(6,4))
plt.boxplot(df["Weight_kg"])
plt.title("Box Plot of Weight")
plt.show()
```

```python
                       #Bar Chart (Categorical Variables)
#Gender
plt.figure(figsize=(6,4))
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

#Country
plt.figure(figsize=(8,5))
df["Country"].value_counts().plot(kind="bar")
plt.title("Country Distribution")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()

#Drug Name
plt.figure(figsize=(10,5))
df["Drug_Name"].value_counts().plot(kind="bar")
plt.title("Drug Distribution")
plt.xlabel("Drug")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
                                                 #Count Plot (Seaborn)
#Gender
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Gender")
plt.title("Gender Count")
plt.show()

#Country
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Country")
plt.xticks(rotation=45)
plt.title("Country Count")
plt.show()

#Drug Name
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Drug_Name")
plt.xticks(rotation=45)
plt.title("Drug Count")
plt.show()
```

```python
                                                      #Pie chart
#Gender
plt.figure(figsize=(6,6))
df["Gender"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.title("Gender Distribution")
plt.show()

#Outcome
plt.figure(figsize=(6,6))
df["Outcome"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.title("Outcome Distribution")
plt.show()
```

<h1>Bivariate Analysis<h1>

```python
#Scatter Plot (Age vs Weight)
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Weight_kg"])
plt.title("Age vs Weight")
plt.xlabel("Age")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
#Correlation Matrix
df.select_dtypes(include=["number"]).corr()
```

```python
#Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(
    df.select_dtypes(include=["number"]).corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#Grouped Bar Chart
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="Gender", y="Age")
plt.title("Average Age by Gender")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(data=df, x="Gender", y="Weight_kg")
plt.title("Average Weight by Gender")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Reporter_Type", y="Age")
plt.title("Average Age by Reporter Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#Cross Tabulation
pd.crosstab(df["Gender"], df["Serious"])
```

```python
#Stacked Bar Chart
pd.crosstab(df["Gender"], df["Serious"]).plot(
    kind="bar",
    stacked=True,
    figsize=(6,4))
plt.title("Gender vs Serious Cases")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Box Plot by Category
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Gender", y="Age")
plt.title("Age Distribution by Gender")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Gender", y="Weight_kg")
plt.title("Weight Distribution by Gender")
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x="Gender", y="Age")
plt.title("Age Distribution by Gender")
plt.show()
```
